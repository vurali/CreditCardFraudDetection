# Credit Card Fraud Detection

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Dataset link : https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud?resource=download&select=creditcard.csv

```python
# Load the data into Pandas DataFrame
data = pd.read_csv('creditcard.csv')
```

```python
# First 5 rows of the data
data.head()
```

```python
# Last 5 rows of the data
data.tail()
```

```python
# Data information
data.info()
```

```python
# Checking the null values in each column
data.isnull().sum()
```

```python
# Distribution of the legit transactions and fraudulent transactions
data.Class.value_counts()
```

This dataset was highly unbalanced.

0 --> Noraml Transaction

1 --> Fraud Transaction

```python
# Separating the data for analysis
legit = data[data.Class == 0]
fraud = data[data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# Statistical measure of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the both transactions
data.groupby('Class').mean()
```

Under-Samplaing

Build a Sample dataset containing similar distribution of Noraml transactions and Fradulent transactions

Number of Fradulent transaction --> 492

```python
legit_sample = legit.sample(n=492)
```

Concatenation two DataFrames

```python
sample_dataset = pd.concat([legit_sample, fraud], axis = 0)
sample_dataset.head()
```

```python
sample_dataset.tail()
```

```python
sample_dataset.Class.value_counts()
```

```python
sample_dataset.groupby('Class').mean()
```

Spliting the data into features and targets

```python
X = sample_dataset.drop(columns='Class', axis=1)
Y = sample_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Spliting the data into Traning data and Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
# Tranin the Logistic Regression model with training dataset
model.fit(X_train, Y_train)
```

Model Evalution

Accuracy Score

```python
# Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print(f'Accuracy on training data : {training_data_accuracy}')
```

```python
# Accuracy on testing data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print(f'Accuracy on testing data: {test_data_accuracy}')
```
